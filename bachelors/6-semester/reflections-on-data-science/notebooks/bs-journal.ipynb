```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


# Data
years = ["2024 Total Steel\nProduction (Mt)", "Total Steel\nProduction 2025 (Mt)"]
values = [80.7, 81.8]
colors = ["#FF0000", "#2ECC71"]

fig, ax = plt.subplots(figsize=(10, 7))

bars = ax.bar(years, values, color=colors, width=0.5, zorder=2)

# Value labels on top of bars
for bar, val in zip(bars, values):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.3,
        str(val),
        ha="center",
        va="bottom",
        fontsize=14,
        fontweight="bold",
    )

# Axes styling
ax.set_ylim(0, 90)
ax.set_yticks(range(0, 91, 10))
ax.tick_params(axis="y", labelsize=12)
ax.tick_params(axis="x", labelsize=13)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

# Title
ax.set_title(
    "Total U.S. Steel Production (Mt) 2024 v. 2025",
    fontsize=18,
    fontweight="bold",
    pad=20,

)

plt.tight_layout()
plt.savefig("steel_production_chart.png", dpi=150, bbox_inches="tight")
plt.show()
```
