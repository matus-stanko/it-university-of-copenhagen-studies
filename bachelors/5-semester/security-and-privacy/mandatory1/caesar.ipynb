```python
def caesar_shift(s: str, shift: int) -> str:
    out = []
    for ch in s:
        if 'a' <= ch <= 'z':
            out.append(chr((ord(ch) - 97 + shift) % 26 + 97))
        elif 'A' <= ch <= 'Z':
            out.append(chr((ord(ch) - 65 + shift) % 26 + 65))
        else:
            out.append(ch)
    return ''.join(out)

msg = "byffixulehymmgsifxzlcyhx"

print("All possible Caesar shifts:\n")
for k in range(26):
    print(f"Shift {k:2d}: {caesar_shift(msg, k)}")
```

```python
# simple_bruteforce.py
import hashlib

salt = "dzr0fv3hz4"
target = "20048def4d4a3922a933a4f84b9cb838c70a8b80402ac45243ed586dd8643bca".lower()

# prechádzame 000000 .. 999999 (vrátane leading zeros)
for i in range(1_000_000):
    pw = f"{i:06d}"                         # napr. "000042"
    h = hashlib.sha256((pw + salt).encode()).hexdigest()
    if h == target:
        print("FOUND password:", pw)
        break
else:
    print("Password not found in 6-digit numeric space.")
```

```python
import json
import pandas as pd

# load pôvodný JSON
with open("full.json", "r") as f:
    data = json.load(f)

# nová štruktúra
filtered = {}

for key, values in data.items():
    sha_items = [item for item in values if item.get("ioc_type") == "sha256_hash"]
    if sha_items:
        filtered[key] = sha_items

# extrahovanie len ioc_value do DataFrame
ioc_values = []
for values in filtered.values():
    for item in values:
        ioc_values.append(item["ioc_value"])

df_filtered = pd.DataFrame(ioc_values, columns=["ioc_value"])

df_filtered
```

```python
# inner join with df_filtered on ioc_value and df on sha256
merged = pd.merge(df_filtered, df, left_on="ioc_value", right_on="sha256", how="inner")
merged.to_csv("files_filtered.csv", index=False)
```
